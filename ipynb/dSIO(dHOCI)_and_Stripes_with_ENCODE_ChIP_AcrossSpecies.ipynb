```r
r.eN.io <- rbind(r.eNL %>% mutate(logQ=-log10(q)) %>% dplyr::select(-c2,-s2,-e2,-ID1,-ID2) %>%
      set_names("c","s","e","q","logQ"),
      r.eNL %>% mutate(logQ=-log10(q)) %>% dplyr::select(-c1,-s1,-e1,-ID1,-ID2) %>%
      set_names("c","s","e","q","logQ")) %>%
group_by(c,s,e) %>% 
summarise(sumFDR=sum(logQ),
          N=n()) %>% 
arrange(sumFDR) 
r.eN.io$rank <- 1:dim(r.eN.io)[1]
```

```r
r.RG.io <- rbind(r.RGL %>% mutate(logQ=-log10(q)) %>% dplyr::select(-c2,-s2,-e2,-ID1,-ID2) %>%
      set_names("c","s","e","q","logQ"),
      r.RGL %>% mutate(logQ=-log10(q)) %>% dplyr::select(-c1,-s1,-e1,-ID1,-ID2) %>%
      set_names("c","s","e","q","logQ")) %>%
group_by(c,s,e) %>% 
summarise(sumFDR=sum(logQ),
          N=n()) %>% 
arrange(sumFDR) 
r.RG.io$rank <- 1:dim(r.RG.io)[1]
```

```r
r.tmp.gr <- dplyr::full_join(r.RG.io,r.eN.io,by=c("c","s","e")) %>%
select(-rank.x,-rank.y) %>%
replace(is.na(.),0) %>%
set_names("seqnames","start","end","RG.interactive","N.rg","EN.interactive","N.en") %>% 
makeGRangesFromDataFrame(.,keep.extra.columns = T)
```

```r
r.EN.dSIO.res.sio <- cbind(r.tmp[queryHits(findOverlaps(r.tmp.gr,rhesus.SIO.gr)),-c(1:3)],
rhesus.SIO[subjectHits(findOverlaps(r.tmp.gr,rhesus.SIO.gr)),]) %>%
group_by(seqnames,start,end) %>%
summarise(RG.interactive=sum(RG.interactive),
          N.rg=sum(N.rg),
          EN.interactive=sum(EN.interactive),
          N.en=sum(N.en))
```

```r
r.EN.dSIO.res <- r.tmp[queryHits(findOverlaps(r.tmp.gr,rhesus.SIO.gr)),] %>% 
as.data.frame()%>% distinct()
```

```r
wilcox.test(r.EN.dSIO.res.sio$EN.interactive,r.EN.dSIO.res.sio$RG.interactive,alternative = "greater")
```
